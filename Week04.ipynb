```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
im_gray = cv2.imread('flower.jpg', cv2.IMREAD_GRAYSCALE)
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
im_bw = cv2.threshold(im_gray, thresh, 255, cv2.THRESH_BINARY)[1]
cv2.imwrite('bw_flower.jpg',im_bw)
plt.imshow(im_bw, cmap='gray')
```

```python
im = cv2.imread('bw_flower.jpg')
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(im,kernel,iterations = 1)
cv2.imshow("image",erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(erosion)
```

```python
im = cv2.imread('bw_flower.jpg')
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(im,kernel,iterations = 1)
cv2.imshow("image",dilation)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(dilation)
```

```python
im = cv2.imread('bw_flower.jpg')
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(im, cv2.MORPH_OPEN,kernel)
cv2.imshow("image",erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(opening)
```

```python
im = cv2.imread('bw_flower.jpg')
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(im, cv2.MORPH_CLOSE,kernel)
cv2.imshow("image",erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(closing)
```

```python
im = cv2.imread('bw_flower.jpg')
kernel = np.ones((5,5),np.uint8)
tophat = cv2.morphologyEx(im, cv2.MORPH_TOPHAT,kernel)
cv2.imshow("image",erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(tophat)
```

```python
im = cv2.imread('bw_flower.jpg')
kernel = np.ones((5,5),np.uint8)
blackhat = cv2.morphologyEx(im, cv2.MORPH_BLACKHAT,kernel)
cv2.imshow("image",erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(blackhat)
```

```python
from skimage import morphology, filters
from skimage import data
import matplotlib.image as mpimg
```

```python
im = mpimg.imread('bw_flower.jpg')
binary = im > filters.threshold_otsu(im)
np.unique(binary)
```

```python
skeleton = morphology.skeletonize(binary)
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(12,12), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(im)
ax[0].axis('off')
ax[0].set_title('original', fontsize=20)

ax[1].imshow(skeleton)
ax[1].axis('off')
ax[1].set_title('skeleton', fontsize=20)

fig.tight_layout()
plt.imshow(skeleton,cmap='gray')
```

```python
thinned = morphology.thin(binary)
thinned_partial = morphology.thin(binary, max_iter=25)
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(12,12), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(im)
ax[0].axis('off')
ax[0].set_title('original', fontsize=20)

ax[1].imshow(thinned)
ax[1].axis('off')
ax[1].set_title('thinned', fontsize=20)

fig.tight_layout()
plt.imshow(thinned,cmap='gray')
```

```python
fig,axes = plt.subplots(2, 2, figsize=(8,8), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(binary, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('original', fontsize=20)

ax[1].imshow(skeleton, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('skeleton', fontsize=20)

ax[2].imshow(thinned, cmap=plt.cm.gray)
ax[2].axis('off')
ax[2].set_title('thinned', fontsize=20)

ax[3].imshow(thinned_partial, cmap=plt.cm.gray)
ax[3].axis('off')
ax[3].set_title('partially thinned', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
convexhull = morphology.convex_hull_image(binary)
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(12,12), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(im, cmap='gray')
ax[0].axis('off')
ax[0].set_title('original', fontsize=20)

ax[1].imshow(convexhull)
ax[1].axis('off')
ax[1].set_title('convexhull', fontsize=20)

fig.tight_layout()
plt.imshow(convexhull,cmap='gray')
```
