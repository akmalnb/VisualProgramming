```python
import matplotlib.pylab as plt
import numpy as np
import cv2
```

```python
im1 = cv2.imread('flower1.jpg',0)
hist = cv2.calcHist([im1],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
hist,bins = np.histogram(im1.ravel(),256,[0,256])
#hist = np.bincount(im1.ravel(),minlength=256)
plt.plot(hist)
```

```python
im1 = cv2.imread('flower1.jpg')
color = ('b','g','r')
for i,col in enumerate (color):
    histr = cv2.calcHist([im1],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
