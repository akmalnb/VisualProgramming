# Library

```python
from PIL import Image, ImageFont, ImageDraw
import requests
from io import BytesIO
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Jawaban

```python
#Url to image
url = "https://images.unsplash.com/photo-1455894127589-22f75500213a?ixid=MXwxMjA3fDB8MHxzZWFyY2h8OHx8bGFwdG9wJTIwb24lMjBkZXNrfGVufDB8fDB8&ixlib=rb-1.2.1&w=1000&q=80"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

#Create font in image
font = ImageFont.truetype("arial.ttf", 25)
d = ImageDraw.Draw(img)
d.rectangle((350, 300, 650, 550), fill= "cyan")
d.text((355, 410), text="Akmal Nusa Bhakti-31187", fill="white", font=font)

#Save image
img.save("tugas.jpeg", "JPEG")
plt.imshow(img)
img.show()
```

```python
#Make some blurred image
im3 = cv2.imread ('tugas.jpeg')

blur = cv2.blur(im3,(55,55))

plt.subplot(121),plt.imshow(im3),plt.title('Original')
#plt.xticks([]), plt,yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])

plt.show(img)
```
